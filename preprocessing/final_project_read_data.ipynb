```python
import pandas as pd
import numpy as np
import os
```

```python
# create a dictionary that maps each sample to its description
mappings = pd.read_csv('mappings.csv', header=None, names=['sample', 'description'])
mappings = dict(zip(mappings['sample'], mappings['description']))
```

```python
all_files = [f for f in os.listdir("GSE140440_RAW") if f.endswith('.gz')]

lst = []
for filename in all_files:
    path = 'GSE140440_RAW/'+ filename
    # read in values from sample
    df = pd.read_csv(path, compression='gzip', header=None, sep='\t')
    df = df.T
    df.columns = df.iloc[0]
    df = df[1:]
    # add columns including description of the sample
    sample = filename.split("_")[0]
    descriptions = mappings[sample]
    cell_line, response, ID = descriptions.split("_")
    df.insert(loc=0, column='ID', value=ID)
    df.insert(loc=0, column='response', value=response)
    df.insert(loc=0, column='cell_line', value=cell_line)
    df.insert(loc=0, column='sample', value=sample)
    lst.append(df)
# merge into one dataframe
data = pd.concat(lst, axis=0, ignore_index=True)
```

```python
# save data to csv for future use
data.to_csv('all_data.tsv', sep='\t')
```
