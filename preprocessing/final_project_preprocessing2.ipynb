```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
# read in dataframe
df = pd.read_csv('all_data.tsv', sep='\t', index_col=0)
```

```python
# change names
# check that all items have only one | (no output expected)
og_gene_names = list(df.columns[4:])
for gene in og_gene_names:
    if len(gene.split("|"))!=2:
        print(gene.split("|"))
```

```python
df.columns.values[4:] = [gene.split("|")[1] for gene in list(df.columns[4:])]
df
```

```python
# separate by PC and DU
df_pc = df[df['cell_line']=='PC3']
df_pc.reset_index(inplace=True)
df_pc = df_pc.rename(columns = {'index':'og_df_idx'})
df_du = df[df['cell_line']=='DU145']
df_du.reset_index(inplace=True)
df_du = df_du.rename(columns = {'index':'og_df_idx'})
```

```python
# function for filtering genes
def get_highest_var(df,num_data_idx_start,pctl_cutoff):
    """
    df: original dataframe
    num_data_idx_start: the column index at which numerical data of interest starts
    pctl_cutoff: only genes at or above this percentile will be considered. e.g. 99.5
    
    Filters by the variance per COLUMN.
    """
    gene_vars = df.iloc[:,num_data_idx_start:].var()
    high_vars = (gene_vars >= np.percentile(gene_vars,pctl_cutoff))
    excl_genes = high_vars.keys()[~high_vars]
    return df[df.columns[~df.columns.isin(excl_genes)]]
```

```python
du_filtered = get_highest_var(df_du,5,99.5)
pc_filtered = get_highest_var(df_pc,5,99.5)
```

```python
# histogram of logged variances across all genes
# just have two plots with translucent overlaid histograms
fig, axs = plt.subplots(2,2,figsize=(30,20))
axs[0,0].hist(np.log(np.var(df_pc.iloc[:,5:],axis=0)+1))
axs[0,0].set_xlabel("Log of Variance")
axs[0,0].set_ylabel("Count")
axs[0,0].set_title("All Genes PC3 Cell Line")
axs[0,1].hist(np.log(np.var(df_du.iloc[:,5:],axis=0)+1))
axs[0,1].set_xlabel("Log of Variance")
axs[0,1].set_ylabel("Count")
axs[0,1].set_title("All Genes DU145 Cell Line")

axs[1,0].hist(np.log(np.var(pc_filtered.iloc[:,5:],axis=0)))
axs[1,0].set_xlabel("Log of Variance")
axs[1,0].set_ylabel("Count")
axs[1,0].set_title("Filtered Genes PC3 Cell Line")
axs[1,1].hist(np.log(np.var(du_filtered.iloc[:,5:],axis=0)))
axs[1,1].set_xlabel("Log of Variance")
axs[1,1].set_ylabel("Count")
axs[1,1].set_title("Filtered Genes DU145 Cell Line");
```

```python
# normalize filtered data
def log2norm(df,num_data_idx_start):
    zscores = zscore(np.log2(df.iloc[:,num_data_idx_start:]+1),axis=0)
    return pd.concat([df.iloc[:,0:5],zscores],axis=1)
```

```python
du_filtnorm = log2norm(du_filtered,5)
pc_filtnorm = log2norm(pc_filtered,5)
```
